```python
import requests
import time

data = []

for i in range(101):
    print(i)
    api_ping = requests.get("http://api.open-notify.org/iss-now.json")
    json_data = api_ping.json()
    data.append(json_data)
    time.sleep(10)
```

```python
data
```

```python
body = []

for record in data:
        row = []
        row.append(record["timestamp"])
        row.append(record["iss_position"]["latitude"])
        row.append(record["iss_position"]["longitude"])
        body.append(row)
```

```python
body
```

```python
import matplotlib.pyplot as plt

latitudes = []
longitudes = []
timestamps = []

timestamps = [row[0] for row in body]
latitudes = [row[1] for row in body]
longitudes = [row[2] for row in body]

fig = plt.figure(figsize=(20, 20))

plt.scatter(longitudes, latitudes)

for i, label in enumerate(timestamps):
    plt.text(longitudes[i], latitudes[i], label, ha='center', va='bottom')

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Path of the International Space Station")
plt.grid(True)

plt.show()
```
